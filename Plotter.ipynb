```python
import pandas as pd
import matplotlib.pyplot as plt
import datetime
%pylab inline
pylab.rcParams['figure.figsize']=(12,8)
#%matplotlib notebook
```

```python
data = pd.read_csv('2900128278.csv', delimiter=';', index_col='recorded', parse_dates=True)
data.head()
```

```python
data.describe()
```

```python
fig,ax = plt.subplots()
ax.plot(data.index, data['radonShortTermAvg (pCi/L)'])
ax.set_ylabel("Radon Short Term Average (pCi/L)")
```

```python
fig,ax = plt.subplots()
ax.plot(data.index, data['radonShortTermAvg (pCi/L)'])
ax.set_ylabel("Radon Short Term Average (pCi/L)")
ax.set_title("Hourly Radon Sampling October 2018")
ax.set_xlim([datetime.date(2018, 10, 1), datetime.date(2018, 11, 1)])
ax.set_ylim(0, 9)
```

```python
fig,ax = plt.subplots()
ax.plot(data.index, data['radonShortTermAvg (pCi/L)'])
ax.set_ylabel("Radon Short Term Average (pCi/L)")
ax.set_title("Hourly Radon Sampling December 2018")
ax.set_xlim([datetime.date(2018, 12, 1), datetime.date(2019, 1, 1)])
ax.set_ylim(0, 9)
```
